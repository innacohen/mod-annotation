```python
import os
import requests
import pandas as pd

def download_mod(url, save_path):
    try:
        r = requests.get(url)
        r.raise_for_status()
        with open(save_path, "wb") as f:
            f.write(r.content)
        return True
    except Exception as e:
        print(f"[ERROR] Failed to download {url}: {e}")
        return False

json_df_sample = pd.read_json("mod_files_sample.json")
```

```python
def View(df, rows=None, cols=None, width=None):
    """Displays the first `rows` of the DataFrame like R's View() by adjusting Pandas settings."""
    
    # Show only the first `rows` of the DataFrame
    with pd.option_context(
        "display.max_rows", rows,  # Limit number of rows shown
        "display.max_columns", cols,  # Show all columns
        "display.max_colwidth", width,  # Show full column width
        "display.expand_frame_repr", False  # Prevent column wrapping
    ):
        display(df.head(rows))  # Show only the first `rows`
```

```python
# Filter rows where the content includes "TITLE Sodium persistent current"
matches = json_df_sample[json_df_sample['content'].str.contains(r'TITLE\s+CA1 KM channel', case=False, na=False)]

# Display the matching rows
View(matches[['row_id', 'file_hash', 'download_url', 'content']].head())
```

```python
output_dir = "../mod_downloads"
os.makedirs(output_dir, exist_ok=True)

for i, row in json_df_sample.iterrows():
    url = row['download_url']
    fname = f"mod_{i}.mod"
    fpath = os.path.join(output_dir, fname)

    if download_mod(url, fpath):
        print(f"[✓] Downloaded to {fpath}")
    else:
        print(f"[✗] Skipped {url}")
```
