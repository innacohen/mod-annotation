```python
from _utils import *

# Set up logging for errors only
LOG_FILE_FP = os.path.join(log_dir, f"nmodl_download_errors_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log")
logging.basicConfig(
    filename=LOG_FILE,
    level=logging.ERROR,  # Only log errors
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# Counter for successful and failed downloads
successful_count = 0
failed_count = 0

# Load metadata
METADATA_FP = 
print(f"Loading metadata from {metadata_path}")
with open(metadata_path, "r") as file:
    raw_data = json.load(file)
    
# Create dataframe from raw data
raw_json_df = pd.DataFrame(raw_data["entries"])

# Filter for annotated samples
ANNOTATED_SAMPLES = pd.read_excel("../annotations/model_db_annotations.xlsx").query("annotated=='y'")["file_hash"].tolist()
print(f"Found {len(ANNOTATED_SAMPLES)} annotated samples")

# Process the dataframe
json_df2 = (
    raw_json_df
    .query("file_hash in @ANNOTATED_SAMPLES")
    .assign(
        has_include = lambda df: df["content"].apply(lambda x: 1 if get_include(x) else 0)
    )
)
print(f"Processing {len(json_df2)} annotated MOD files")

# Set up output directories
base_dir = "../data/raw/nmodl"
includes_dir = "../data/raw/nmodl/includes"
os.makedirs(base_dir, exist_ok=True)
os.makedirs(includes_dir, exist_ok=True)

# Download the MOD files
for _, row in tqdm(json_df2.iterrows(), total=len(json_df2), desc="Downloading MOD files"):
    try:
        url = row["download_url"]
        file_hash = row["file_hash"]
        
        # Download and save as filehash.mod
        response = requests.get(url)
        response.raise_for_status()
        
        output_path = os.path.join(base_dir, f"{file_hash}.mod")
        with open(output_path, "wb") as f:
            f.write(response.content)
        
        # Increment success counter
        successful_count += 1
        
    except Exception as e:
        error_msg = f"Error downloading {url}: {e}"
        logging.error(f"File hash: {file_hash}, URL: {url} - {error_msg}")
        print(error_msg)
        
        # Increment failure counter
        failed_count += 1

# Log and print completion statistics
stats_message = f"\nTotal files processed: {len(json_df2)}, " \
               f"Successfully downloaded: {successful_count}, " \
               f"Failed downloads: {failed_count}"
print(stats_message)

if failed_count > 0:
    logging.error(stats_message)
    print(f"Some downloads failed. Check log file: {LOG_FILE}")
else:
    print("All downloads completed successfully")
```
