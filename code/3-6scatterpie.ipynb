```python
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge
import numpy as np

def pie_scatter(points, radius=0.15, ax=None):
    """
    Plot (x, y) points where each point is rendered as a mini pie chart.

    Parameters
    ----------
    points : list of tuples
        Each tuple: (x, y, color_list)
        Example: (4, 2, ["orange", "blue"])
    radius : float
        Radius of each mini pie
    ax : matplotlib axis (optional)

    Returns
    -------
    ax : matplotlib axis
    """

    if ax is None:
        fig, ax = plt.subplots()

    for x, y, colors in points:
        n = len(colors)

        # Single-color point: draw a full circle
        if n == 1:
            wedge = Wedge((x, y), radius, 0, 360, facecolor=colors[0], edgecolor="black")
            ax.add_patch(wedge)

        # Multi-color: split into equal wedges
        else:
            angles = np.linspace(0, 360, n + 1)
            for i in range(n):
                wedge = Wedge(
                    (x, y),
                    radius,
                    angles[i],
                    angles[i + 1],
                    facecolor=colors[i],
                    edgecolor="black"
                )
                ax.add_patch(wedge)

    # Autoscale so patches are visible
    ax.autoscale_view()
    ax.set_aspect("equal")

    return ax


# --------------------------
# Example usage
# --------------------------

points = [
    (4, 2, ["orange", "blue"]),
    (1, 1, ["purple"]),
    (0, 0, ["orange", "blue", "purple"]),
    (2, 3, ["orange"]),
    (3, 1, ["blue", "purple"]),
]

fig, ax = plt.subplots(figsize=(6, 6))

pie_scatter(points, radius=0.2, ax=ax)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Mini-Pie Scatter Plot")

plt.show()
```
